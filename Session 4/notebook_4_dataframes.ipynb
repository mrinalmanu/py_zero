```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
# create a scaler object
```

```python
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
df_scaled
```

```python
corr = df_scaled.corr()
px.imshow(corr, text_auto=True, aspect='auto')
```

```python
# let's look at all the variables that showed a larger than 10% correlation 
# we use abs or absolute because we want to look at both positive and negative ends
corr[abs(corr['DEATH_EVENT']) > 0.1]['DEATH_EVENT']
```

https://scikit-learn.org/stable/

```python
from sklearn.model_selection import train_test_split
```

```python
# Trial and Error revealed that not considering Age column improves accuracy

x = df_scaled[['ejection_fraction', 'serum_creatinine', 'serum_sodium', 'time']]


'''
x = (x-x.mean())/x.std() 

We don't need to perform this step, because we already scaled the data
'''

y = df['DEATH_EVENT']

#Spliting data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, 
                                                    y,
                                                    random_state=1,
                                                    test_size=0.2)
```

https://scikit-learn.org/stable/modules/svm.html

```python
from sklearn.svm import SVC
svm=SVC()
svm??
```

```python
svm.fit(x_train,y_train)
```

```python
# now we use our model to predict the outcomes
p4=svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
s4=accuracy_score(y_test,p4)
print("Support Vector Machine Success Rate :", "{:.2f}%".format(100*s4))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, p4, labels=svm.classes_)
```

```python
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=svm.classes_)
```
